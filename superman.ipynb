```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("SuperMan.jpg")
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
background_img = cv2.imread("sky.jpg")
background_img_RGB = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)

plt.imshow(background_img_RGB)
```

```python
rows, cols, _ = background_img.shape()
img = cv2.resize(img, (cols, rows))

img_R = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_H = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(img_H)

for i in range(rows):
    for j in range(cols):
        if 25 < H[i, j] < 80 and S[i, j] > 60 and V[i, j] > 70:
            img_R[i, j] = background_img_RGB[i, j]
            
plt.imshow(img_R)
```
